```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data.drop(['Address'],axis=1,inplace=True)
data.head()
```

###### EDA (Exploratory data analysis)

```python
sns.pairplot(data, x_vars=['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population'], y_vars='Price',aspect=0.9,height=9)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
x=data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=data['Price']
```

```python
x
```

```python
y
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.fit(x_test,y_test)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
train_predict=model.predict(x_train)
```

```python
test_predict=model.predict(x_test)
```

###### evaluation metrics

```python
train_res=y_train - train_predict
train_res
```

```python
test_res = y_test- test_predict
test_res
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
